# Handling the missing values

```python
data["last review"]=pd.to_datetime(data["last review"], errors ="coerce")

```

```python
data.info()
```

```python
# There are many columns which has showing missing values in it , so to correct it , we will using  customize fillna
```

```python
data.fillna({"reviews per month":0 ,"last review":data["last review"].min()}, inplace=True)
```

```python
data.isna().sum()
```

```python
# there 250 missing values in NAME column and 406 missing valus in host name in data set , we will drop that mising valus 
```

```python
data.dropna(subset= ["NAME","host name"] , inplace=True)
```

```python
data.isna().sum()
```

```python
# here we are droping the column name license and  house rules by using drop syntax 
```

```python
data=data.drop(columns=["license","house_rules"],errors="ignore")
```

```python
data.head()
```

```python
#  Initially there were 26 columns but after dropping 2 columns , it is now 24 columns in tha  data 
```

```python
#  If we see the price and service fee columns in given data set , there is symbol of dollor($) and it creates error , so we remove that symbol from that columns by data cleaning process 

#  data type of price and service fee column are in object (means string) , we have to convert into float(numerical values)
```
